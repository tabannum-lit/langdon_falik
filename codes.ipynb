```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

df_falik_1= pd.read_csv('/content/drive/MyDrive/Probha/FalikDataset.csv')
df_langdon_1= pd.read_csv('/content/drive/MyDrive/Probha/LangdonDataset.csv')
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(7,5))
im=ax1.scatter(x=df_langdon_1['CO3'], y=df_langdon_1['G'])
ax1.set(ylabel='Clarification rate/(mmol CACO_3/m^2.day)', xlabel='[CO_3^2-](mu mol/kg of sweater)')
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(7,5))
im=ax1.plot(df_langdon_1['CO3'], df_langdon_1['G'])
ax1.set(ylabel='Clarification rate/(mmol CACO_3/m^2.day)', xlabel='[CO_3^2-](mu mol/kg of sweater)')
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(7,5))
sns.barplot(data=df_falik_1, x='plant', y='width', hue='type', palette=['b', 'r'])
ax1.set(ylabel='Stomatal Opening (mu)', xlabel='Plant Number')
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(7,5))
sns.boxplot(data=df_falik_1, x="type", y="width")
ax1.set(ylabel='Stomatal Opening (mu)', xlabel='', xticks=['control', '15min'], xticklabels=['control', 'treatment'])
```
